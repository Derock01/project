# Which month maximum number of forest fires were reported and visualize them


```python
 max_month = df.groupby('month')['number'].sum().idxmax()

# Visualize monthly forest fires
plt.figure(figsize=(12, 6))
sns.barplot(x='month', y='number', data=df, estimator=sum, ci=None, order=df['month'].unique())
plt.title(f'Monthly Forest Fires - Maximum in {max_month}')
plt.show()
```

# Which Year maximum number of forest fires were reported and visualize them

```python
max_year = df.groupby('year')['number'].sum().idxmax()

# Visualize yearly forest fires
plt.figure(figsize=(12, 6))
sns.barplot(x='year', y='number', data=df, estimator=sum, ci=None, order=df['year'].unique())
plt.title(f'Yearly Forest Fires - Maximum in {max_year}')
plt.show()
```

# Which states maximum number of forest fires were reported and visualize

# Find the total number of fires reported in the Amazonas



```python
 amazonas_total_fires = df[df['state'] == 'Amazonas']['number'].sum()
```

# Display number of fires reported in Amazonas (year-wise) and visualize 



```python
amazonas_yearly_fires = df[df['state'] == 'Amazonas'].groupby('year')['number'].sum()

plt.figure(figsize=(12, 6))
sns.barplot(x=amazonas_yearly_fires.index, y=amazonas_yearly_fires.values)
plt.title('Number of Fires Reported in Amazonas (Year-wise)')
plt.show()
```

# Display number of fires reported in Amazonas (day-wise) and visualize 



```python
amazonas_daily_fires = df[df['state'] == 'Amazonas'].groupby('date')['number'].sum()

plt.figure(figsize=(12, 6))
amazonas_daily_fires.plot()
plt.title('Number of Fires Reported in Amazonas (Day-wise)')
plt.show()
```

# Find the total number of fires that were reported in 2015 and visualize based on each month



```python
fires_2015 = df[df['year'] == 2015].groupby('month')['number'].sum()

plt.figure(figsize=(12, 6))
sns.barplot(x=fires_2015.index, y=fires_2015.values)
plt.title('Number of Fires Reported in 2015 (Month-wise)')
plt.show()
```

# Average number of fires reported from highest to lowest(state-wise) and visualize


```python
avg_fires_by_state = df.groupby('state')['number'].mean().sort_values(ascending=False)

plt.figure(figsize=(14, 6))
sns.barplot(x=avg_fires_by_state.index, y=avg_fires_by_state.values)
plt.title('Average Number of Fires Reported (State-wise)')
plt.show()
```
